## UTA

```python
import pulp

def uta_method_with_binary_inconsistency_solver(alternatives, criteria, weights, thresholds):
    # Initialize the linear programming problem
    prob = pulp.LpProblem("UTA Method with Binary Inconsistency Solver", pulp.LpMaximize)

    # Create decision variables for each alternative and criterion
    decision_vars = pulp.LpVariable.dicts("Alternatives", [(alt, crit) for alt in alternatives for crit in criteria], 
                                          lowBound=0, upBound=1, cat=pulp.LpInteger)

    # Create the objective function
    prob += pulp.lpSum([weights[i] * pulp.lpSum([decision_vars[(alt, i)] for alt in alternatives]) for i in range(len(weights))])

    # Create constraints for each criterion
    for i, criterion in enumerate(criteria):
        for alt in alternatives:
            prob += decision_vars[(alt, criterion)] >= thresholds[i][0] - (1 - decision_vars[(alt, criterion)]) * 1000
            prob += decision_vars[(alt, criterion)] <= thresholds[i][1] + (1 - decision_vars[(alt, criterion)]) * 1000

    # Solve the linear programming problem
    prob.solve()

    # Print the results
    if prob.status == pulp.LpStatusOptimal:
        print("The optimal solution is:")
        for alt in alternatives:
            print(f"{alt}: {pulp.value(pulp.lpSum([decision_vars[(alt, i)] * weights[i] for i in range(len(weights))]))}")
    else:
        print("The problem is infeasible or unbounded.")
```

```python
weights = [0.51, 0.28, 0.08, 0.08, 0.05]

aesthetics_threshold = 6
commutation_threshold = 30
gym_num_threshold = 3
```

```python
data_norm = alternatives.apply(lambda x: x/x.max(), axis=0)
```

```python
consistency = pulp.LpVariable.dicts("Consistency", ((i, j) for i in data_norm.index for j in data_norm.columns), 0, 1, pulp.LpBinary)

prob = pulp.LpProblem("UTA", pulp.LpMaximize)

prob += pulp.lpSum([consistency[(i, j)] for i in data_norm.index for j in data_norm.columns])

# Add the consistency constraints
for i in data_norm.index:
    for j in data_norm.columns:
        if j == 'aesthetics':
            prob += consistency[(i, j)] == (data_norm.loc[i, j] >= aesthetics_threshold)
        elif j == 'commutation':
            prob += consistency[(i, j)] == (data_norm.loc[i, j] >= commutation_threshold)
        elif j == 'gym_num':
            prob += consistency[(i, j)] == (data_norm.loc[i, j] >= gym_num_threshold)
        else:
            prob += consistency[(i, j)] == 1

# Add the UTA weights constraints
prob += pulp.lpSum([weights[0]*data_norm.loc[i, 'price']*consistency[(i, 'price')] + 
               weights[1]*data_norm.loc[i, 'area']*consistency[(i, 'area')] +
               weights[2]*data_norm.loc[i, 'aesthetics']*consistency[(i, 'aesthetics')] +
               weights[3]*data_norm.loc[i, 'commutation']*consistency[(i, 'commutation')] + weights[4]*data_norm.loc[i, 'gym_num']*consistency[(i, 'gym_num')]
               for i in data_norm.index]) == 1
```

```python
prob.solve()

print("Status:", pulp.LpStatus[prob.status])

if pulp.LpStatus[prob.status] == "Optimal":
    print("Solution:")
    for i in data_norm.index:
        print(f"Property {i+1}: {alternatives.loc[i, 'price']} (Price) + {alternatives.loc[i, 'area']} (Area) + {alternatives.loc[i, 'aesthetics']} (Aesthetics) + {alternatives.loc[i, 'commutation']} (Commutation) + {alternatives.loc[i, 'gym_num']} (Gym Num) = {pulp.value(prob.objective)}")
else:
    print("No optimal solution found.")
```
