## UTA

Get minimal $\alpha$ and maximal $\beta$ values for all criteria

```python
alternatives = get_alternatives()

def get_min_values(alternatives):
    mins = alternatives.min()
    # if column is int change value to int
    mins_dict = {}
    for col in alternatives.columns:
        if alternatives[col].dtype == np.int64:
            mins_dict[col] = int(mins[col])
        else:
            mins_dict[col] = mins[col]
    return mins_dict

def get_max_values(alternatives):
    maxs = alternatives.max()
    # if column is int change value to int
    maxs_dict = {}
    for col in alternatives.columns:
        if alternatives[col].dtype == np.int64:
            maxs_dict[col] = int(maxs[col])
        else:
            maxs_dict[col] = maxs[col]

    return maxs_dict
get_min_values(alternatives), get_max_values(alternatives)
```

Estimation constraints

```python
import pulp

prob = pulp.LpProblem("UTA_Method_with_Binary_Inconsistency_Solver", pulp.LpMinimize)

def estimation_constraints(number_of_alternatives, problem):

    # each alternative has over and under estimation error variable
    over = pulp.LpVariable.dicts("over", range(number_of_alternatives))
    under = pulp.LpVariable.dicts("under", range(number_of_alternatives))

    # minimize the sum of over and under estimation error
    problem += pulp.lpSum([over[i] + under[i] for i in range(number_of_alternatives)])

    for i in range(number_of_alternatives):
        problem += over[i] >= 0
        problem += under[i] >= 0
    return problem

prob = estimation_constraints(len(alternatives), prob)
# prob
```

Unique value variables

```python

def criterion_values_variables(alternatives):

    # variable for each unique value of each criterion
    vars = dict()
    for criterion in alternatives.columns:
        unique_values = alternatives[criterion].unique()
        for value in unique_values:
            vars[criterion + "_" + str(value)] = pulp.LpVariable(criterion + "_" + str(value))

    return vars

vars = criterion_values_variables(alternatives)
vars
```

Normalization:

$u_i(\alpha_i) = 0$


$\sum{u_i(\beta_i)} = 1$


Monotonicity:

$u_i(x_i^{j+1}) - u_i(x_i^j) \ge 0$ 

```python
def criterion_values_constraints(alternatives, vars, problem, gains):
    # nonnegativity constraints
    for var in vars.values():
        problem += var >= 0

    # get maximum and minimum values of each criterion
    min_values = get_min_values(alternatives)
    max_values = get_max_values(alternatives)
    alpha = dict()
    beta = dict()
    for criterion in alternatives.columns:
        if gains[criterion]:
            alpha[criterion] = min_values[criterion]
            beta[criterion] = max_values[criterion]
        else:
            alpha[criterion] = max_values[criterion]
            beta[criterion] = min_values[criterion]

    # each criterion for min value is 0
    for criterion in alternatives.columns:
        problem += vars[criterion + "_" + str(alpha[criterion])] == 0
    
    # sum of max value criteria is 1
    
    problem += pulp.lpSum([vars[criterion + "_" + str(beta[criterion])] for criterion in alternatives.columns]) == 1

    # monotonicity constraints
    for criterion in alternatives.columns:
        vals = alternatives[criterion].unique()
        # sort values in ascending order if gain, descending if not gain
        if gains[criterion]:
            vals = sorted(vals)
        else:
            vals =sorted(vals,reverse=True)

                
        for i in range(len(vals) - 1):
            problem += vars[criterion + "_" + str(vals[i])] <= vars[criterion + "_" + str(vals[i + 1])]

    return problem

gains = {"price": False, "area": True, "aesthetics": True, "commutation": False, "gym_num": True}

prob = criterion_values_constraints(alternatives, vars, prob, gains)
```

Pairwise comparisons

```python
# function to change ranking into set of pairwise comparisons

def ranking_to_comparisons(ranking):
    comparisons = []
    for i in range(len(ranking)-1):
        comparisons.append((ranking[i],ranking[i+1]))
    return comparisons
comparisons = ranking_to_comparisons(ranking)
comparisons
```

```python
def pairwise_comparisons(prob,vars,alternatives,comparisons):

    for left,right in comparisons:
        left_side = [vars[criterion + "_" + str(alternatives[criterion][left])] for criterion in alternatives.columns]
        left_side.extend([prob.variablesDict()["over_" + str(left)], prob.variablesDict()["under_" + str(left)]])
        right_side = [vars[criterion + "_" + str(alternatives[criterion][right])] for criterion in alternatives.columns]
        right_side.extend([prob.variablesDict()["over_" + str(right)], prob.variablesDict()["under_" + str(right)]])
        
        equation = pulp.lpSum(left_side) >= pulp.lpSum(right_side) + 0.01
        print(equation)
        prob += equation
    return prob

prob = pairwise_comparisons(prob,vars,alternatives,comparisons)
```

```python
# plot marginal values of criteria
import matplotlib.pyplot as plt
import numpy as np

def plot_marginal_values(vars):
    # get unique values of each criterion
    unique_values = dict()
    for criterion in alternatives.columns:
        unique_values[criterion] = alternatives[criterion].unique()
        unique_values[criterion].sort()

    # get marginal values of each criterion
    marginal_values = dict()
    for criterion in alternatives.columns:
        marginal_values[criterion] = [vars[criterion + "_" + str(value)].varValue for value in unique_values[criterion]]

    # plot marginal values
    # subplots for each criterion
    fig, axs = plt.subplots(len(alternatives.columns), 1, figsize=(10, 10))
    for i, criterion in enumerate(alternatives.columns):
        axs[i].plot(unique_values[criterion], marginal_values[criterion])
        # increase margin between plots
        plt.subplots_adjust(hspace=0.5)
        axs[i].set_xlabel(criterion)
        axs[i].set_ylabel("Marginal Value")
        axs[i].grid()
    plt.show()
        

plot_marginal_values(prob.variablesDict())
```

```python
# calculate ranking based on the values of the variables
def calculate_ranking(alternatives, vars):

    values = dict()

    for i in range(len(alternatives)):
        values[i] = 0
        
        for c in alternatives.columns:
            values[i] += vars[c + "_" + str(alternatives[c][i])].varValue
    print(values)
    return sorted(values, key=values.get, reverse=True)

ranking_2 = calculate_ranking(alternatives, prob.variablesDict())
print(ranking,ranking_2)
```

## Finding inconsistencies


```python
#comparisons = ranking_to_comparisons(ranking)
comparisons = []
comparisons.extend([(12,9),(9,6),(6,12),(6,9),(9,12)])
# 12 > 9 > 6 ; 6 > 12 ; 6 > 9
comparisons
```

```python
# for each pairwise comparison add binary constraint


def find_inconsistencies(alternatives,comparisons,gains):
    
    criteria_vars = criterion_values_variables(alternatives)
    problem = pulp.LpProblem("inconsistency",sense = pulp.LpMinimize)
    
    problem = criterion_values_constraints(alternatives,criteria_vars,problem,gains)
    vars = []
    for left, right in comparisons:
        var = pulp.LpVariable(f"var_{left}_{right}",cat=pulp.const.LpBinary)
        vars.append(var)
        #left_side
        left_side = [criteria_vars[criterion + "_" + str(alternatives[criterion][left])] for criterion in alternatives.columns]
        right_side = [criteria_vars[criterion + "_" + str(alternatives[criterion][right])] for criterion in alternatives.columns]
        problem += left_side >= right_side - var + 0.0001
    problem += pulp.lpSum(vars)

    problem.solve()
    def resolve_inconsistencies(problem,vars):

        inconsistencies = []
        for var in vars:
            if var.varValue == 1:
                #print(var.name, var.varValue)
                inconsistencies.append(var)
        problem += pulp.lpSum(inconsistencies) <= len(inconsistencies) -1
        problem.solve()
        return problem,inconsistencies
        
    subsets = []
    problem, inconsistencies = resolve_inconsistencies(problem,vars)
    subsets.append(inconsistencies)
    
    while problem.status > 0:
            problem, inconsistencies = resolve_inconsistencies(problem,vars)
            subsets.append(inconsistencies)

   # print(subsets)
    return subsets
            

subsets = find_inconsistencies(alternatives,comparisons,gains)

subsets
```

```python
def remove_subset(comparisons,subset):
    for var in subset:
        split = var.name.split("_")
        greater, lesser = split[1],split[2]
        #print((int(greater),int(lesser)))
        comparisons.remove((int(greater),int(lesser)))
    return comparisons

remove_subset(comparisons, subsets[0])
```

## Final Algorithm

```python
def uta(alternatives,gains,ranking):
    comparisons = ranking_to_comparisons(ranking)
    subsets = find_inconsistencies(alternatives,comparisons,gains)
    if len(subsets[0]) > 0:
        print("Removing subsets")
        print(subsets) 
        remove_subset(comparisons, subsets[0])

    prob = pulp.LpProblem("UTA_Method_with_Binary_Inconsistency_Solver", pulp.LpMinimize)
    prob = estimation_constraints(len(alternatives), prob)
    vars = criterion_values_variables(alternatives)

    gains = {"price": False, "area": True, "aesthetics": True, "commutation": False, "gym_num": True}

    prob = criterion_values_constraints(alternatives, vars, prob, gains)
    
    prob = pairwise_comparisons(prob,vars,alternatives,comparisons)

    prob.solve()
    print(prob)
    
    plot_marginal_values(prob.variablesDict())

    ranking_2 = calculate_ranking(alternatives, prob.variablesDict())
    print(ranking_2)

alternatives = get_alternatives()
gains = {"price": False, "area": True, "aesthetics": True, "commutation": False, "gym_num": True}
ranking = [9,12,13,3,11,0,4,5,2,8,10,7,1,6]
#ranking = [12,13,9,0,11,3,4,5,2,8,10,7,1,6]
#ranking = sorted(ranking)
uta(alternatives,gains,ranking)
```

```python
def pairwise_matrix(rankings):
    n = len(rankings)
    matrix = [[0]*n for i in range(n)]
    for i in range(n):
        for j in range(i+1, n):
            if rankings.index(i) < rankings.index(j):
                matrix[i][j] = 1
            else:
                matrix[j][i] = 1
    return matrix
```

```python
def kendalls_tau(matrix1, matrix2):
    kendalls_dist = 0
    for row1, row2 in zip(matrix1, matrix2):
        for i in range(len(row1)):
            kendalls_dist += abs(row1[i] - row2[i])
    kendalls_dist *= 0.5
    tau = 1-4*kendalls_dist/(len(matrix1)*(len(matrix1)-1))
    return tau
```
